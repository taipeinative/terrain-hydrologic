```python
import numpy as np
import matplotlib.pyplot as plt

_dx = 1

x_axis = np.arange(0, 100, _dx, dtype = float)
z_axis = np.zeros(x_axis.shape)
z_axis[x_axis > 50] += 50
new_z_axis = z_axis.copy()

print(z_axis)
```

```python
ErosionParam = 0.1
Period = 1000 # Years
_dt = 0.2 * _dx * _dx / ErosionParam

for i in range(int(Period / _dt)):

    _qs = -ErosionParam * np.diff(new_z_axis) / _dx
    _dzdt = -np.diff(_qs) / _dx
    new_z_axis[1:-1] += _dzdt * _dt

fig, ax = plt.subplots()
ax.plot(x_axis, z_axis, c = 'grey', label = 'Original')
ax.plot(x_axis, new_z_axis, c = 'red', label = 'Diffused')
ax.set(title = f'1D Scarp Evolution after {Period} Years')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

x_axis = np.arange(0, 100, _dx, dtype = float)
# z_axis = np.zeros(x_axis.shape)
# z_axis[x_axis > 50] += 50

z_axis = np.zeros(x_axis.shape)
z_axis[x_axis > 20] += 20
z_axis[x_axis > 60] += 40
z_axis[x_axis > 90] += 10

ax.plot(x_axis, z_axis, c = 'grey', ls = 'dotted', label = 'Original')

for d in np.arange(0.01, 0.11, 0.01):

    new_z_axis = z_axis.copy()
    _dt = 0.2 * _dx * _dx / d

    for i in range(int(Period / _dt)):
        _qs = -d * np.diff(new_z_axis) / _dx
        _dzdt = -np.diff(_qs) / _dx
        new_z_axis[1:-1] += _dzdt * _dt

    ax.plot(x_axis, new_z_axis, label = f'D = {round(d, 2)}')

ax.set(title = f'1D Scarp Evolution after {Period} Years with different Ds')

ax.set(xlabel = 'Distance (x)', ylabel = 'Elevation (z)')
ax.legend()
plt.show()
```
