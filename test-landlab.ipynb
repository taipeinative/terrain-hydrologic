# Landlab Test

A simple Landlab tutorial taken from [official manual](https://landlab.readthedocs.io/en/latest/user_guide/index.html) website. You can use them to check your Landlab conditions and learn basic usages.

```python
from landlab import RasterModelGrid
from landlab.plot.graph import plot_graph

grid = RasterModelGrid((4, 5), 1)
plot_graph(grid, at = 'node')
plot_graph(grid, at = 'link')
plot_graph(grid, at = 'cell')
```
